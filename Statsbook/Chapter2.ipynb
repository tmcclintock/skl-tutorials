```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
%matplotlib inline
```

```python
#2.3
def plot(p):
    x = np.arange(0,7)
    dist = binom.pmf(x, 6, p)
    plt.plot(x, dist)
plot(0.5)
plot(1./6.)
```

```python
#2.4
def P(x):
    if x < 7:
        return (x-1)/36.
    elif x >= 7:
        return (13-x)/36.
x = np.arange(2,13)
Px = np.array([P(xi) for xi in x])
mean = np.sum(x*Px)
std = np.sqrt(np.sum(x*x*Px)-mean**2)
print mean, std, x[len(x)/2]
```
